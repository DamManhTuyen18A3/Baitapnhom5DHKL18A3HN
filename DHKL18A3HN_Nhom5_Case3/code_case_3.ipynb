```python
import numpy as np
print("Đọc dữ liệu campaign_A.csv")
data_A = np.genfromtxt("campaign_A.csv", delimiter=",", dtype=None, names=True, encoding='utf-8')

dates_A = data_A['date']    
visits_A = data_A['visits'] 

print("Campaign A - Dates:", dates_A)
print("Campaign A - Visits:", visits_A) 
print("Đọc dữ liệu từ campaign_B.csv")

data_B = np.genfromtxt("campaign_B.csv", delimiter=",", dtype=None, names=True, encoding='utf-8')

dates_B = data_B['date']
visits_B = data_B['visits']

print("Campaign B - Dates:", dates_B)
print("Campaign B - Visits:", visits_B)
dates_A = dates_A.astype('datetime64[D]')
dates_B = dates_B.astype('datetime64[D]')
```

```python
print("Chuyển cột ngày sang định dạng datetime64[D]: ")
dates_A = dates_A.astype('datetime64[D]')
dates_B = dates_B.astype('datetime64[D]')
```

```python
print("5 ngày đầu và cuối của campaign_A:")
print(dates_A.dtype, visits_A.dtype)
print( dates_A.shape, visits_A.shape)

print("5 ngày đầu của campaign_A: ")
print(data_A[0:5])

print("5 ngày cuối của campaign_A: ")
print( data_A[-5:])
```

```python
print("5 ngày đầu và cuối của campaign_B:")
print(dates_B.dtype, visits_B.dtype)
print( dates_B.shape, visits_B.shape)

print("5 ngày đầu của campaign_B: ")
print(data_B[0:5])

print("5 ngày cuối của campaign_B: ")
print( data_B[-5:])
```

```python
bay_ngay_dau = data_A[0:7]
bay_ngay_cuoi = data_A[-7:]
ngay_giua =len (data_A)//2
bay_ngay_giua =data_A[ngay_giua-3 : ngay_giua+4]

print('Bay ngay dau:\n',bay_ngay_dau)
print('Bay ngay cuoi:\n',bay_ngay_cuoi)
print('Bay ngay giua:\n',bay_ngay_giua)
```

```python
bay_ngay_dau = data_B[0:7]
bay_ngay_cuoi = data_B[-7:]
ngay_giua =len (data_B)//2
bay_ngay_giua =data_B[ngay_giua-3 : ngay_giua+4]

print('Bay ngay dau:\n',bay_ngay_dau)
print('Bay ngay cuoi:\n',bay_ngay_cuoi)
print('Bay ngay giua:\n',bay_ngay_giua)
```

```python
mask_A_lon_hon_B = visits_A > visits_B
print('\n',mask_A_lon_hon_B)

cac_ngay_A_hon_B = dates_A[mask_A_lon_hon_B]
print('Cac ngay Campaign A co luong truy cap cao hon Campaign B la:\n',cac_ngay_A_hon_B)
```

```python
mask_tang_truong_dot_bien_B= visits_B[1:] - visits_B[:-1]
luong_truy_cap_dot_bien = 1000
X = mask_tang_truong_dot_bien_B > 1000
date_check = dates_B[1:]
cac_ngay_tang_truong_dot_bien = date_check[X]
print('So cac ngay Campaign B co luong truy cap tang truong dot bien la:\n',cac_ngay_tang_truong_dot_bien)
```

```python
idx_sorted = np.argsort(visits_A)
top5_idx = idx_sorted[-5:]

top5_dates = dates_A[top5_idx]
top5_visits = visits_A[top5_idx]
print("Top 5 ngày có lượng truy cập cao nhất của campiagn_A là :")
for date, visit in zip(top5_dates, top5_visits):
    print(date, visit)
```

```python
idx_sorted = np.argsort(visits_B)
top5_idx = idx_sorted[-5:]

top5_dates = dates_B[top5_idx]
top5_visits = visits_B[top5_idx]
print("Top 5 ngày có lượng truy cập cao nhất của campiagn_B là :")
for date, visit in zip(top5_dates, top5_visits):
    print(date, visit)
```

```python
chenh_lech_tung_ngay_giua_hai_chien_dich = visits_A - visits_B
chenh_lech_tung_ngay_giua_hai_chien_dich
chenh_lech = np.column_stack((dates_A.astype(str), chenh_lech_tung_ngay_giua_hai_chien_dich))
print('Chenh lech tung ngay giua chien dich A so voi B la:\n',chenh_lech)
```

```python
A_diff_day = np.diff(visits_A)
B_diff_day = np.diff(visits_B)
print('Phan tich bien dong ngay lien ke cua chien dich A: ', A_diff_day)
print('Phan tich bien dong ngay lien ke cua chien dich B: ', B_diff_day)
```

```python
mean_v = np.mean(visits_A)
std_v = np.std(visits_A)
max_v = np.max(visits_A)
min_v = np.min(visits_A)
argmax_v = np.argmax(visits_A)
argmin_v = np.argmin(visits_A)
    
print("Thống kê cơ bản campaign_A: ")
print(f"Mean: {mean_v}")
print(f"Std: {std_v}")
print(f"Max: {max_v} (ngày: {dates_A[argmax_v]})")
print(f"Min: {min_v} (ngày: {dates_A[argmin_v]})")
print(f"Argmax index: {argmax_v}")
print(f"Argmin index: {argmin_v}\n")
```

```python
mean_v = np.mean(visits_B)
std_v = np.std(visits_B)
max_v = np.max(visits_B)
min_v = np.min(visits_B)
argmax_v = np.argmax(visits_B)
argmin_v = np.argmin(visits_B)
    
print("Thống kê cơ bản campaign_B: ")
print(f"Mean: {mean_v}")
print(f"Std: {std_v}")
print(f"Max: {max_v} (ngày: {dates_B[argmax_v]})")
print(f"Min: {min_v} (ngày: {dates_B[argmin_v]})")
print(f"Argmax index: {argmax_v}")
print(f"Argmin index: {argmin_v}\n")
```

```python
tong_hop = np.column_stack((dates_A.astype(str), visits_A.astype(str), visits_B.astype(str),chenh_lech_tung_ngay_giua_hai_chien_dich.astype(str)))
np.savetxt("ket_qua_case_3.csv", tong_hop, fmt= "%s", delimiter = ",", header = "Dates, VisitsA, VisitB, Diff")
```
