```python
# 1. Đọc campaign_A.csv và campaign_B.csv bằng np.loadtxt hoặc np.genfromtxt.
import numpy as np
data_A = np.genfromtxt("campaign_A.csv", delimiter=",", dtype=None, names=True) 
data_B = np.genfromtxt("campaign_B.csv", delimiter=",", dtype=None, names=True)
```

```python
# 2. Tách cột ngày và cột lượng truy cập thành hai mảng riêng.
# Campaign A
print("CAMPAIGN A")
dates_A = data_A['date']    
visits_A = data_A['visits'] 
print("Campaign A - Dates:", dates_A)
print("Campaign A - Visits:", visits_A)
print("\n")
# Campaign B 
print("CAMPAIGN B")
dates_B = data_B['date']
visits_B = data_B['visits']
print("Campaign B - Dates:", dates_B)
print("Campaign B - Visits:", visits_B)
```

```python
# 3. Chuyển cột ngày sang định dạng datetime64[D]. 
dates_A = dates_A.astype('datetime64[D]')
dates_B = dates_B.astype('datetime64[D]')
```

```python
# 4. Kiểm tra dtype, shape, và in ra 5 giá trị đầu – cuối của mỗi chiến dịch. 
# Campaign A
print("CAMPAIGN A")
print(dates_A.dtype, visits_A.dtype)
print( dates_A.shape, visits_A.shape)

print("5 ngay dau cua campaign_A: ")
print(data_A[0:5])

print("5 ngay cuoi cua campaign_A: ")
print( data_A[-5:])
print("\n")
# Campaign B
print("CAMPAIGN B")
print(dates_B.dtype, visits_B.dtype)
print( dates_B.shape, visits_B.shape)

print("5 ngay dau cua campaign_B: ")
print(data_B[0:5])

print("5 ngay cuoi cua campaign_B: ")
print( data_B[-5:])
```

```python
# 5. Thực hiện slicing: lấy 7 ngày đầu, 7 ngày giữa và 7 ngày cuối của mỗi chiến dịch.
# Campaign A
print("CAMPAIGN A") 
bay_ngay_dau = data_A[0:7]
bay_ngay_cuoi = data_A[-7:]
ngay_giua =len (data_A)//2
bay_ngay_giua =data_A[ngay_giua-3 : ngay_giua+4]

print('Bay ngay dau:\n',bay_ngay_dau)
print('Bay ngay cuoi:\n',bay_ngay_cuoi)
print('Bay ngay giua:\n',bay_ngay_giua)
print("\n")
# Campaign B
print("CAMPAIGN B")
bay_ngay_dau = data_B[0:7]
bay_ngay_cuoi = data_B[-7:]
ngay_giua =len (data_B)//2
bay_ngay_giua =data_B[ngay_giua-3 : ngay_giua+4]

print('Bay ngay dau:\n',bay_ngay_dau)
print('Bay ngay cuoi:\n',bay_ngay_cuoi)
print('Bay ngay giua:\n',bay_ngay_giua)
```

```python
# 6. Boolean mask: tìm các ngày Campaign A có lượng truy cập cao hơn Campaign B. 
mask_A_lon_hon_B = visits_A > visits_B
print('\n',mask_A_lon_hon_B)
print("\n")
cac_ngay_A_hon_B = dates_A[mask_A_lon_hon_B]
print('Cac ngay Campaign A co luong truy cap cao hon Campaign B la:\n',cac_ngay_A_hon_B)
```

```python
# 7. Boolean mask: tìm các ngày Campaign B tăng trưởng đột biến so với ngày trước đó. 
mask_tang_truong_dot_bien_B= visits_B[1:] - visits_B[:-1]
luong_truy_cap_dot_bien = 1000
X = mask_tang_truong_dot_bien_B > 1000
date_check = dates_B[1:]
cac_ngay_tang_truong_dot_bien = date_check[X]
print('So cac ngay Campaign B co luong truy cap tang truong dot bien la:\n',cac_ngay_tang_truong_dot_bien)
```

```python
# 8. Fancy indexing: tìm top-5 ngày có lượng truy cập cao nhất của mỗi chiến dịch. 
top5_A_idx = np.argsort(visits_A)[-5:]
top5_B_idx = np.argsort(visits_B)[-5:]

top5_A = dates_A[top5_A_idx], visits_A[top5_A_idx]
top5_B = dates_B[top5_B_idx], visits_B[top5_B_idx]

print('Top 5 ngay co luong truy cap cao nhat o chien dich A la: \n',top5_A)
print("\n")
print('Top 5 ngay co luong truy cap cao nhat o chien dich B la: \n',top5_B)
```

```python
# 9. Broadcasting: tính chênh lệch truy cập từng ngày giữa hai chiến dịch. 
chenh_lech_tung_ngay_giua_hai_chien_dich = visits_A - visits_B
chenh_lech_tung_ngay_giua_hai_chien_dich
chenh_lech = np.column_stack((dates_A.astype(str), chenh_lech_tung_ngay_giua_hai_chien_dich))
print('Chenh lech tung ngay giua chien dich A so voi B la:\n',chenh_lech)
```

```python
# 10. Dùng np.diff() để phân tích biến động truy cập ngày liền kề của từng chiến dịch. 
A_diff_day = np.diff(visits_A)
B_diff_day = np.diff(visits_B)
print('Phan tich bien dong ngay lien ke cua chien dich A:\n ', A_diff_day)
print("\n")
print('Phan tich bien dong ngay lien ke cua chien dich B:\n', B_diff_day)
```

```python
# 11. Tính các thống kê cơ bản: mean, std, max, min, argmax, argmin cho mỗi chiến dịch.
# Campaign A
print("CAMPAIGN A")
mean_v = np.mean(visits_A)
std_v = np.std(visits_A)
max_v = np.max(visits_A)
min_v = np.min(visits_A)
argmax_v = np.argmax(visits_A)
argmin_v = np.argmin(visits_A)
    
print("Thong ke co ban campaign_A: ")
print(f"Mean: {mean_v}")
print(f"Std: {std_v}")
print(f"Max: {max_v} (ngay: {dates_A[argmax_v]})")
print(f"Min: {min_v} (ngay: {dates_A[argmin_v]})")
print(f"Argmax index: {argmax_v}")
print(f"Argmin index: {argmin_v}\n")

# Campaign B
print("CAMPAIGN B")
mean_v = np.mean(visits_B)
std_v = np.std(visits_B)
max_v = np.max(visits_B)
min_v = np.min(visits_B)
argmax_v = np.argmax(visits_B)
argmin_v = np.argmin(visits_B)
    
print("Thong ke co ban campaign_B: ")
print(f"Mean: {mean_v}")
print(f"Std: {std_v}")
print(f"Max: {max_v} (ngay: {dates_B[argmax_v]})")
print(f"Min: {min_v} (ngay: {dates_B[argmin_v]})")
print(f"Argmax index: {argmax_v}")
print(f"Argmin index: {argmin_v}\n")
```

```python
# 12. Tạo bảng tổng hợp và lưu ket_qua_case_3.csv bằng np.savetxt. 
tong_hop = np.column_stack((dates_A.astype(str), visits_A.astype(str), visits_B.astype(str),chenh_lech_tung_ngay_giua_hai_chien_dich.astype(str)))
np.savetxt("ket_qua_case_3.csv", tong_hop, fmt= "%s", delimiter = ",", header = "Dates, VisitsA, VisitB, Diff")
```
